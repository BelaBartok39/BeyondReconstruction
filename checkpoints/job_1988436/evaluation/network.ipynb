```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
class SimpleNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(in_features=784, out_features=128)
        self.fc2 = nn.Linear(in_features=128, out_features=10)

    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
    
        return x
```

```python
model = SimpleNet()

# The Judge (Loss Function)
criterion = nn.CrossEntropyLoss()

# The Mechanic (Optimizer)
optimizer = optim.SGD(model.parameters(), lr=0.01)

print("🤖 Model initialized. Starting training on random noise...")
```

```python
for step in range(101):
    # A. Create dummy data (Batch Size 64)
    # 64 random images, flattened to 784 pixels each
    inputs = torch.randn(64, 784)
    # 64 random labels (integers between 0 and 9)
    labels = torch.randint(0, 10, (64,))

    # B. Reset gradients
    optimizer.zero_grad()

    # C. Forward Pass (Guess)
    outputs = model(inputs)

    # D. Calculate Loss (Score)
    loss = criterion(outputs, labels)

    # E. Backward Pass (Diagnosis)
    loss.backward()

    # F. Optimization Step (Fix)
    optimizer.step()

    # Print progress every 20 steps
    if step % 20 == 0:
        print(f"Step {step}: Loss = {loss.item():.4f}")

print("\n✅ Training complete!")
```

```python
print("\n🔮 Running inference on a single fake image...")
model.eval() # Switch to evaluation mode

# Create 1 fake image
fake_image = torch.randn(1, 784)

with torch.no_grad():
    logits = model(fake_image)

# Calculate probabilities and prediction
probs = F.softmax(logits, dim=1)
predicted_digit = torch.argmax(logits, dim=1)

print(f"Raw Logits: \n{logits}")
print(f"Probabilities: \n{probs}")
print(f"Predicted Digit: {predicted_digit.item()}")
```
